```python
# Drop unnecessary columns:
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Strip whitespace in string columns
string_cols = ['Product', 'Purchase Address', 'City']
for col in string_cols:
    df[col] = df[col].str.strip()
```

```python
# Validate 'Sales' column
df['Calculated Sales'] = df['Quantity Ordered'] * df['Price Each']

# Identify mismatches (rounded to 2 decimal places)
sales_mismatch = df[df['Calculated Sales'].round(2) != df['Sales'].round(2)]

# If mismatches found, fix them and notify
if not sales_mismatch.empty:
    print(f"Found {len(sales_mismatch)} mismatched 'Sales' values. Updating them...")

    # Update only the mismatched rows
    df.loc[sales_mismatch.index, 'Sales'] = df.loc[sales_mismatch.index, 'Calculated Sales']
else:
    print("No mismatched 'Sales' values found.")
```

```python
# Extract State from 'Purchase Address'
df['State'] = df['Purchase Address'].apply(lambda x: x.split(',')[2].split()[0] if ',' in x else None)
```

```python
#dropping duplicated values:
df= df.drop_duplicates(subset=['Order ID','Product','Quantity Ordered','Order Date','Purchase Address'])
```

```python
# To get time
df['Time'] =df['Order Date'].str.split(' ').str.get(1)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = False)
```

```python
# changing orderdate to datetime dtype
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = False)
```

```python
# converting to day time formate
df['Time'] = pd.to_datetime(df['Time'],format = '%H:%M:%S')
df['time'] = df['Time'].dt.time
```

```python
# final cleaned dataset after removing duplicated values
print("\nCleaned Data Summary:")
print(df.info())
```
